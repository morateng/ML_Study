```python
train_x = train.drop('voted', axis=1)
train_y = train['voted']
```

```python
train_x['age_group']=1
train_x['gender']=1
train_x['race']=1
train_x['religion']=1

test['age_group']=1
test['gender']=1
test['race']=1
test['religion']=1
```

```python
model = lgbm.LGBMClassifier(n_estimators=500)
model.fit(train_x, train_y)
```
